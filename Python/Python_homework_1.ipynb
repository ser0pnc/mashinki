```python
 def task_05_func(text: str) -> dict:
    list_of_words = sorted(text.split(' '))

    set_of_words = set(list_of_words)

    dict_of_words = dict()

    for word in set_of_words:
        dict_of_words[word] = list_of_words.count(word)

    return dict_of_words
```

```python
def task_06_func(input_str: str, input_char: str) -> tuple:

    list_of_indexes = list()

    for index, char in enumerate(input_str):
        if char == input_char:
            list_of_indexes.append(index)

    if len(list_of_indexes) >= 2:
        return (list_of_indexes[0], list_of_indexes[-1])
    elif len(list_of_indexes) == 1:
        return (list_of_indexes[0], None)
    else:
        return (None, None)
```

```python
def task_07_func(lst: list) -> list:

    list_without_negatives = list(filter(lambda x: x >= 0, lst))

    list_of_squares = list(map(lambda x: x ** 2, list_without_negatives))

    return sorted(list_of_squares, reverse=True)
```

```python
def task_08_func(lst: list, index: int) -> iter:
    return iter(sorted(lst, key=lambda x: x[index].lower(), reverse=True))


def random_strings(n, k):
    import numpy as np 

    result = list()

    english_letters_low = [chr(ord('a') + i) for i in range(26)]

    english_letters_up = list(map(lambda x: x.upper(), english_letters_low))
    english_letters = np.array(english_letters_low + english_letters_up)


    for i in range(n):
        indexes = list(np.random.randint(0, len(english_letters), k))
        
        # result.append(''.join(english_letters[indexes]))
        result.append(tuple(english_letters[indexes]))

    return result
```
